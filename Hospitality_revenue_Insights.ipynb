```python
# Hospitality Revenue Insights Dashboard
import pandas as pd
import matplotlib.pyplot as plt

#  Sample Data Creation
data = {
    'Date': pd.date_range(start='2025-01-01', periods=10, freq='D'),
    'Rooms_Available': [100, 100, 100, 100, 100, 100, 100, 100, 100, 100],
    'Rooms_Sold': [80, 75, 90, 95, 85, 70, 60, 100, 92, 88],
    'Room_Revenue': [80000, 75000, 95000, 105000, 85000, 70000, 60000, 115000, 92000, 88000],
    'F&B_Revenue': [20000, 18000, 25000, 30000, 22000, 15000, 12000, 32000, 26000, 24000],
    'Other_Revenue': [5000, 4000, 6000, 7000, 5500, 3000, 2500, 8000, 6000, 5000]
}

df = pd.DataFrame(data)

#  Calculating KPIs
df['Occupancy_Rate'] = (df['Rooms_Sold'] / df['Rooms_Available']) * 100
df['ADR'] = df['Room_Revenue'] / df['Rooms_Sold']  # Average Daily Rate
df['RevPAR'] = df['Room_Revenue'] / df['Rooms_Available']  # Revenue per Available Room
df['Total_Revenue'] = df['Room_Revenue'] + df['F&B_Revenue'] + df['Other_Revenue']

#  Display Summary KPIs
summary = {
    'Average Occupancy (%)': round(df['Occupancy_Rate'].mean(), 2),
    'Average ADR': round(df['ADR'].mean(), 2),
    'Average RevPAR': round(df['RevPAR'].mean(), 2),
    'Total Revenue (10 days)': df['Total_Revenue'].sum()
}

print(" --- Hospitality Revenue Insights Summary ---")
for key, value in summary.items():
    print(f"{key}: {value}")

# Visualization

# Occupancy and ADR trends
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Occupancy_Rate'], marker='o', label='Occupancy Rate (%)')
plt.plot(df['Date'], df['ADR'], marker='s', label='ADR ()')
plt.title('Occupancy & ADR Trend')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Revenue breakdown
plt.figure(figsize=(10, 6))
plt.bar(df['Date'], df['Room_Revenue'], label='Room Revenue')
plt.bar(df['Date'], df['F&B_Revenue'], bottom=df['Room_Revenue'], label='F&B Revenue')
plt.bar(df['Date'], df['Other_Revenue'], bottom=df['Room_Revenue'] + df['F&B_Revenue'], label='Other Revenue')
plt.title('Total Revenue Breakdown by Source')
plt.xlabel('Date')
plt.ylabel('Revenue ()')
plt.legend()
plt.show()

# RevPAR trend
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['RevPAR'], color='purple', marker='d', label='RevPAR ()')
plt.title('Revenue per Available Room (RevPAR) Trend')
plt.xlabel('Date')
plt.ylabel('RevPAR ()')
plt.legend()
plt.grid(True)
plt.show()
```
